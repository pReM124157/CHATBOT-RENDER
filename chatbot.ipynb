```python
import re
import math
import base64
import numpy as np
# from mftool import Mftool # Removed as API is no longer used
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from IPython.display import display, Javascript
from google.colab import output
import speech_recognition as sr
import warnings

warnings.filterwarnings("ignore", category=UserWarning)

# Initialize Mutual Fund Scraper (no longer used for dummy data)
# mf = Mftool()

# KNOWLEDGE BASE - Dummy Data for demonstration
KNOWLEDGE_BASE = {
    "sbi bluechip fund": {
        "name": "SBI Bluechip Fund (Direct - Growth)",
        "nav": "₹ 62.50",
        "date": "2024-07-25",
        "min_sip": "₹ 500",
        "risk": "High",
        "category": "Large Cap",
        "details": "यह एक लार्ज कैप इक्विटी फंड है जो मुख्य रूप से बड़ी कंपनियों में निवेश करता है।"
    },
    "hdfc mid-cap opportunities fund": {
        "name": "HDFC Mid-Cap Opportunities Fund (Direct - Growth)",
        "nav": "₹ 165.75",
        "date": "2024-07-25",
        "min_sip": "₹ 1000",
        "risk": "Very High",
        "category": "Mid Cap",
        "details": "यह फंड मध्यम आकार की कंपनियों में निवेश करता है जिनमें विकास की उच्च क्षमता होती है।"
    },
    "parag parikh flexi cap fund": {
        "name": "Parag Parikh Flexi Cap Fund (Direct - Growth)",
        "nav": "₹ 78.90",
        "date": "2024-07-25",
        "min_sip": "₹ 1000",
        "risk": "High",
        "category": "Flexi Cap",
        "details": "यह एक फ्लेक्सी कैप फंड है जो लार्ज, मिड और स्मॉल कैप कंपनियों में निवेश कर सकता है।"
    }
}

# Helper function for number formatting to Lakh/Crore
def format_indian_currency(amount):
    amount = int(amount)
    if amount >= 10000000: # Crore
        return f"₹ {amount / 10000000:.2f} करोड़"
    elif amount >= 100000: # Lakh
        return f"₹ {amount / 100000:.2f} लाख"
    else:
        return f"₹ {amount:,}" # Standard comma separated

# =====================================================================
# MACHINE LEARNING INTENT MATCHING SYSTEM
# =====================================================================
class HindiFinancialNLP:
    def __init__(self):
        # Local Intent Dataset mapping Hindi phrase variations to mathematical actions
        self.dataset = {
            "नमस्ते हेलो हाय नमस्कार हेल्प मदद शुभ प्रभात शुभ संध्या": "greeting",
            "एसआईपी कैलकुलेटर निवेश प्लान रिटर्न एस आई पी एसआईपी गणना एसआईपी की जानकारी": "sip_calc",
            "एसडब्ल्यूपी विड्रॉल बैलेंस निकालें गणना पेंशन मासिक एसडब्ल्यूपी की जानकारी एसडब्ल्यूपी कैसे काम करता है": "swp_calc",
            "एसबीआई एचडीएफसी पराग पारिख फंड एनएवी भाव लाइव म्यूचुअल म्यूचुअल फंड का रेट म्यूचुअल फंड के बारे में बताओ सबसे अच्छी शिप सबसे अच्छा SIP बेस्ट SIP कौन सा म्यूचुअल फंड अच्छा है मेरी आय के अनुसार कौन सा म्यूचुअल फंड अच्छा है कितनी कमाई पर कौन सा फंड लें मेरी सैलरी 20 लाख है कौन सा फंड अच्छा रहेगा बेस्ट फंड कौन सा है": "fund_info"
        }
        # Text vector space representation using a custom word tokenizer
        self.vectorizer = TfidfVectorizer(tokenizer=self.tokenize)
        self.X_train, self.intents = self.train_model()

    def tokenize(self, text):
        """Preprocesses string sequences into standard words"""
        text = text.lower().strip()
        return re.findall(r'\b\w+\b', text, re.UNICODE)

    def train_model(self):
        """Converts strings into structured vector spaces"""
        corpus = list(self.dataset.keys())
        intents = list(self.dataset.values())
        X = self.vectorizer.fit_transform(corpus)
        return X, intents

    def predict(self, text):
        """Uses Cosine Similarity to select matching category matrices"""
        query_vec = self.vectorizer.transform([text])
        similarities = cosine_similarity(query_vec, self.X_train).flatten()
        max_idx = np.argmax(similarities)

        if similarities[max_idx] < 0.10:
            return "unknown"
        return self.intents[max_idx]

# Initialize and fit model instantly
nlp_model = HindiFinancialNLP()

# =====================================================================
# MATHEMATICAL BACKEND & PROCESSING HANDLER
# =====================================================================
def calculate_sip(monthly, rate, years):
    i = (rate / 100) / 12
    n = years * 12
    fv = monthly * (((1 + i)**n - 1) / i) * (1 + i)
    invested = monthly * n
    return round(fv), round(invested), round(fv - invested)

def calculate_swp(principal, withdrawal, rate, years):
    i = (rate / 100) / 12
    months = years * 12
    balance = principal
    withdrawn = 0
    for _ in range(months):
        if balance <= 0:
            break
        balance = (balance * (1 + i)) - withdrawal
        withdrawn += withdrawal
    return max(0, round(balance)), round(withdrawn)

def generate_hindi_response(user_input):
    intent = nlp_model.predict(user_input)
    # Numerical patterns extractor (Slot-Filling)
    numbers = [int(s) for s in re.findall(r'\d+', user_input)]
    user_input_lower = user_input.lower()

    # Guardrail: No stock tips or crypto predictions
    if "stock" in user_input_lower or "शेयर" in user_input_lower or "crypto" in user_input_lower or "क्रिप्टो" in user_input_lower:
        return "Maaf kijiye, main sirf mutual funds aur financial planning se sambandhit jankari de sakta hoon. Stock aur crypto market ki salah main nahi deta."

    if intent == "greeting":
        return ("नमस्ते! मैं **HindiFinancialNLP v2** हूँ, आपका विशेषज्ञ AI वित्तीय सलाहकार।\\n"
                "मैं भारतीय म्यूचुअल फंड, SIP/SWP योजना और व्यक्तिगत वित्त में आपकी मदद कर सकता हूँ।\\n\\n"
                "आप मुझसे ये चीज़ें पूछ सकते हैं:\\n\\n"
                "1. **SIP कैलकुलेटर** (उदा: '5000 रुपये 12% पर 10 साल का SIP')\\n"
                "2. **SWP कैलकुलेटर** (उदा: '1000000 का निवेश 10000 का SWP 8% पर 5 साल')\\n"
                "3. **फंड की जानकारी** (उदा: 'SBI ब्लूचिप फंड का NAV क्या है')")

    elif intent == "sip_calc":
        if len(numbers) >= 3:
            monthly, rate, years = numbers[0], numbers[1], numbers[2]
            fv, invested, returns = calculate_sip(monthly, rate, years)

            # Format numbers using Lakh/Crore
            fv_formatted = format_indian_currency(fv)
            invested_formatted = format_indian_currency(invested)
            returns_formatted = format_indian_currency(returns)

            return (f"📊 **SIP गणना परिणाम:**\\n"
                    f"🔹 हर महीने निवेश: {format_indian_currency(monthly)}\\n"
                    f"🔹 कुल निवेश राशि (Total Invested Amount): {invested_formatted}\\n"
                    f"🔹 अनुमानित मुनाफा (Wealth Gained): {returns_formatted}\\n"
                    f"💰 {years} साल बाद कुल वैल्यू (Total Future Value): {fv_formatted}")
        return "कृपया SIP गणना के लिए मासिक निवेश, अनुमानित दर, और समय (सालों में) स्पष्ट रूप से बताएं।"

    elif intent == "swp_calc":
        if len(numbers) >= 4:
            principal, withdrawal, rate, years = numbers[0], numbers[1], numbers[2], numbers[3]
            bal, withdrawn = calculate_swp(principal, withdrawal, rate, years)

            # Format numbers using Lakh/Crore
            principal_formatted = format_indian_currency(principal)
            withdrawal_formatted = format_indian_currency(withdrawal)
            bal_formatted = format_indian_currency(bal)
            withdrawn_formatted = format_indian_currency(withdrawn)

            return (f"📊 **SWP गणना परिणाम:**\\n"
                    f"🔹 प्रारंभिक कुल निवेश (Initial Investment): {principal_formatted}\\n"
                    f"🔹 मासिक निकासी (Monthly Withdrawal): {withdrawal_formatted}\\n"
                    f"🔹 कुल निकाली गई रकम (Total Withdrawn Amount): {withdrawn_formatted}\\n"
                    f"📉 {years} साल बाद बचा हुआ बैलेंस (Remaining Balance): {bal_formatted}")
        return "कृपया सही आंकड़े प्रदान करें। उदाहरण: 10 लाख का निवेश 10 हज़ार का SWP 8% दर 5 साल।"

    elif intent == "fund_info":
        found_fund = None
        for fund_key, fund_data in KNOWLEDGE_BASE.items():
            if fund_key in user_input_lower:
                found_fund = fund_data
                break

        if found_fund:
            return (f"📈 **फंड जानकारी:**\\n"
                    f"🔹 स्कीम का नाम: {found_fund['name']}\\n"
                    f"🔹 NAV भाव (आज): {found_fund['nav']}\\n"
                    f"🔹 तारीख: {found_fund['date']}\\n"
                    f"🔹 न्यूनतम SIP: {found_fund['min_sip']}\\n"
                    f"🔹 जोखिम स्तर: {found_fund['risk']}\\n"
                    f"🔹 श्रेणी: {found_fund['category']}\\n"
                    f"💬 विवरण: {found_fund['details']}")
        else:
            return ("म्यूचुअल फंड में निवेश आपके वित्तीय लक्ष्यों और जोखिम उठाने की क्षमता पर निर्भर करता है।\\n"
                    "मैं आपको सामान्य जानकारी दे सकता हूँ, लेकिन व्यक्तिगत सलाह के लिए वित्तीय सलाहकार से संपर्क करें।\\n"
                    "अधिक विशिष्ट जानकारी के लिए कृपया फंड का नाम बताएं (जैसे SBI Bluechip, HDFC Mid-Cap, Parag Parikh Flexi Cap).")

    return "Maaf kijiye, mujhe iski jankari nahi hai. Mai sirf verified mutual funds aur calculations me madad kar sakta hoon."

# Create connection handshake across cloud environments
# output.register_callback('notebook.processAudio', process_audio_callback)
```

```python
from flask import Flask, render_template, request, jsonify
from pyngrok import ngrok # Using pyngrok for direct ngrok control
import re
import math
import base64
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import speech_recognition as sr
import warnings
import matplotlib.pyplot as plt
import pandas as pd
from gtts import gTTS
from pydub import AudioSegment
import io
import os
import threading # For running Flask in a separate thread

# Suppress warnings for cleaner output
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

app = Flask(__name__)

# --- Chatbot's KNOWLEDGE BASE (from your notebook) ---
KNOWLEDGE_BASE = {
    "sbi bluechip fund": {
        "name": "SBI Bluechip Fund (Direct - Growth)",
        "nav": "₹ 62.50",
        "date": "2024-07-25",
        "min_sip": "₹ 500",
        "risk": "High",
        "category": "Large Cap",
        "details": "यह एक लार्ज कैप इक्विटी फंड है जो मुख्य रूप से बड़ी कंपनियों में निवेश करता है।"
    },
    "hdfc mid-cap opportunities fund": {
        "name": "HDFC Mid-Cap Opportunities Fund (Direct - Growth)",
        "nav": "₹ 165.75",
        "date": "2024-07-25",
        "min_sip": "₹ 1000",
        "risk": "Very High",
        "category": "Mid Cap",
        "details": "यह फंड मध्यम आकार की कंपनियों में निवेश करता है जिनमें विकास की उच्च क्षमता होती है।"
    },
    "parag parikh flexi cap fund": {
        "name": "Parag Parikh Flexi Cap Fund (Direct - Growth)",
        "nav": "₹ 78.90",
        "date": "2024-07-25",
        "min_sip": "₹ 1000",
        "risk": "High",
        "category": "Flexi Cap",
        "details": "यह एक फ्लेक्सी कैप फंड है जो लार्ज, मिड और स्मॉल कैप कंपनियों में निवेश कर सकता है।"
    }
}

# --- Helper function for number formatting (from your notebook) ---
def format_indian_currency(amount):
    amount = int(amount)
    if amount >= 10000000: # Crore
        return f"₹ {amount / 10000000:.2f} करोड़"
    elif amount >= 100000: # Lakh
        return f"₹ {amount / 100000:.2f} लाख"
    else:
        return f"₹ {amount:,}" # Standard comma separated

# --- MACHINE LEARNING INTENT MATCHING SYSTEM (from your notebook) ---
class HindiFinancialNLP:
    def __init__(self, dataset_list):
        self.dataset = {}
        for item in dataset_list:
            intent = item['intent']
            text = item['text']
            if intent in self.dataset:
                self.dataset[intent] += " " + text
            else:
                self.dataset[intent] = text

        inverted_dataset = {}
        for intent, text_combined in self.dataset.items():
            inverted_dataset[text_combined] = intent
        self.dataset = inverted_dataset

        self.vectorizer = TfidfVectorizer(tokenizer=self.tokenize)
        self.X_train, self.intents = self.train_model()

    def tokenize(self, text):
        text = text.lower().strip()
        return re.findall(r'\b\w+\b', text, re.UNICODE)

    def train_model(self):
        corpus = list(self.dataset.keys())
        intents = list(self.dataset.values())
        X = self.vectorizer.fit_transform(corpus)
        return X, intents

    def predict(self, text):
        query_vec = self.vectorizer.transform([text])
        similarities = cosine_similarity(query_vec, self.X_train).flatten()
        max_idx = np.argmax(similarities)

        if similarities[max_idx] < 0.10:
            return "unknown"
        return self.intents[max_idx]

# Initialize and fit model using the dataset from your notebook
nlp_model = HindiFinancialNLP(financial_nlp_dataset)

# --- MATHEMATICAL BACKEND & PROCESSING HANDLER (from your notebook) ---
def calculate_sip(monthly, rate, years):
    i = (rate / 100) / 12
    n = years * 12
    fv = monthly * (((1 + i)**n - 1) / i) * (1 + i)
    invested = monthly * n

    months_list = []
    invested_list = []
    future_value_list = []
    current_fv = 0
    current_invested = 0

    for month in range(1, n + 1):
        current_invested += monthly
        current_fv = monthly * (((1 + i)**month - 1) / i) * (1 + i)
        months_list.append(month)
        invested_list.append(current_invested)
        future_value_list.append(current_fv)

    plot_df = pd.DataFrame({
        'Month': months_list,
        'Total Invested': invested_list,
        'Future Value': future_value_list
    })

    return round(fv), round(invested), round(fv - invested), plot_df

def calculate_swp(principal, withdrawal, rate, years):
    i = (rate / 100) / 12
    months = years * 12
    balance = principal
    withdrawn = 0
    for _ in range(months):
        if balance <= 0:
            break
        balance = (balance * (1 + i)) - withdrawal
        withdrawn += withdrawal
    return max(0, round(balance)), round(withdrawn)

def generate_hindi_response(user_input):
    intent = nlp_model.predict(user_input)
    numbers = [int(s) for s in re.findall(r'\d+', user_input)]
    user_input_lower = user_input.lower()
    plot_fig = None

    if "stock" in user_input_lower or "शेयर" in user_input_lower or "crypto" in user_input_lower or "क्रिप्टो" in user_input_lower:
        return ("Maaf kijiye, main sirf mutual funds aur financial planning se sambandhit jankari de sakta hoon. Stock aur crypto market ki salah main nahi deta.", None)

    if intent == "greeting":
        # Removed extra parentheses around the multiline string to ensure it's a direct string
        return ("नमस्ते! मैं **HindiFinancialNLP v2** हूँ, आपका विशेषज्ञ AI वित्तीय सलाहकार।\n"
                "मैं भारतीय म्यूचुअल फंड, SIP/SWP योजना और व्यक्तिगत वित्त में आपकी मदद कर सकता हूँ।\n\n"
                "आप मुझसे ये चीज़ें पूछ सकते हैं:\n\n"
                "1. **SIP कैलकुलेटर** (उदा: '5000 रुपये 12% पर 10 साल का SIP')\n"
                "2. **SWP कैलकुलेटर** (उदा: '1000000 का निवेश 10000 का SWP 8% पर 5 साल')\n"
                "3. **फंड की जानकारी** (उदा: 'SBI ब्लूचिप फंड का NAV क्या है')", None)

    elif intent == "sip_calc":
        if len(numbers) >= 3:
            monthly, rate, years = numbers[0], numbers[1], numbers[2]
            fv, invested, returns, plot_df = calculate_sip(monthly, rate, years)

            fv_formatted = format_indian_currency(fv)
            invested_formatted = format_indian_currency(invested)
            returns_formatted = format_indian_currency(returns)

            response_text = (f"📊 **SIP गणना परिणाम:**\n"
                             f"🔹 हर महीने निवेश: {format_indian_currency(monthly)}\n"
                             f"🔹 SIP अवधि में कुल जमा की गई राशि (Total Amount Deposited during SIP Period): {invested_formatted}\n"
                             f"🔹 अनुमानित मुनाफा (Wealth Gained): {returns_formatted}\n"
                             f"💰 {years} साल बाद कुल वैल्यू (Total Future Value): {fv_formatted}")

            fig, ax = plt.subplots(figsize=(12, 7), facecolor='#f0f2f5')
            line_invested, = ax.plot(plot_df['Month'] / 12, plot_df['Total Invested'], label='कुल निवेशित राशि (Total Invested)', color='#5B84B1', linestyle='-', linewidth=2, marker='o', markersize=6, markevery=[-1], markerfacecolor='white', markeredgecolor='#5B84B1')
            line_fv, = ax.plot(plot_df['Month'] / 12, plot_df['Future Value'], label='कुल अनुमानित मूल्य (Total Future Value)', color='#3CB371', linestyle='-', linewidth=2, marker='o', markersize=6, markevery=[-1], markerfacecolor='white', markeredgecolor='#3CB371')
            final_year = plot_df['Month'].iloc[-1] / 12
            ax.annotate(f'₹ {invested_formatted}', xy=(final_year, plot_df['Total Invested'].iloc[-1]), xytext=(5, 5), textcoords='offset points', color='#5B84B1', fontsize=10, fontweight='bold', ha='left')
            ax.annotate(f'₹ {fv_formatted}', xy=(final_year, plot_df['Future Value'].iloc[-1]), xytext=(5, -15), textcoords='offset points', color='#3CB371', fontsize=10, fontweight='bold', ha='left')
            ax.set_title(f'SIP वृद्धि: ₹{monthly}/माह, {rate}% {years} वर्षों के लिए', fontsize=18, color='#333333', pad=20, fontweight='bold')
            ax.set_xlabel('वर्ष (Years)', fontsize=14, color='#555555')
            ax.set_ylabel('राशि (₹)', fontsize=14, color='#555555')
            ax.grid(axis='y', linestyle='-', alpha=0.4, color='#e0e0e0')
            ax.grid(axis='x', visible=False)
            ax.tick_params(axis='x', colors='#666666', labelsize=11)
            ax.tick_params(axis='y', colors='#666666', labelsize=11)
            ax.legend(loc='upper left', frameon=False, fontsize=12, labelcolor=['#5B84B1', '#3CB371'])
            ax.set_facecolor('#ffffff')
            for spine in ax.spines.values():
                spine.set_visible(False)
            plt.tight_layout()
            plot_fig = fig
            plt.close(fig)

            return response_text, plot_fig
        return (" कृपया SIP गणना के लिए मासिक निवेश, अनुमानित दर, और समय (सालों में) स्पष्ट रूप से बताएं।", None)

    elif intent == "swp_calc":
        if len(numbers) >= 4:
            principal, withdrawal, rate, years = numbers[0], numbers[1], numbers[2], numbers[3]
            bal, withdrawn = calculate_swp(principal, withdrawal, rate, years)

            principal_formatted = format_indian_currency(principal)
            withdrawal_formatted = format_indian_currency(withdrawal)
            bal_formatted = format_indian_currency(bal)
            withdrawn_formatted = format_indian_currency(withdrawn)

            return (f"📊 **SWP गणना परिणाम:**\n"
                    f"🔹 प्रारंभिक कुल निवेश (Initial Investment): {principal_formatted}\n"
                    f"🔹 मासिक निकासी (Monthly Withdrawal): {withdrawal_formatted}\n"
                    f"🔹 कुल निकाली गई रकम (Total Withdrawn Amount): {withdrawn_formatted}\n"
                    f"📉 {years} साल बाद बचा हुआ बैलेंस (Remaining Balance): {bal_formatted}", None)
        return (" कृपया सही आंकड़े प्रदान करें। उदाहरण: 10 लाख का निवेश 10 हज़ार का SWP 8% दर 5 साल.", None)

    elif intent == "fund_info":
        found_fund = None
        for fund_key, fund_data in KNOWLEDGE_BASE.items():
            if fund_key in user_input_lower:
                found_fund = fund_data
                break

        if found_fund:
            return (f"📈 **फंड जानकारी:**\n"
                    f"🔹 स्कीम का नाम: {found_fund['name']}\n"
                    f"🔹 NAV भाव (आज): {found_fund['nav']}\n"
                    f"🔹 तारीख: {found_fund['date']}\n"
                    f"🔹 न्यूनतम SIP: {found_fund['min_sip']}\n"
                    f"🔹 जोखिम स्तर: {found_fund['risk']}\n"
                    f"🔹 श्रेणी: {found_fund['category']}\n"
                    f"💬 विवरण: {found_fund['details']}", None)
        else:
            # Removed extra parentheses around the multiline string to ensure it's a direct string
            return ("म्यूचुअल फंड में निवेश आपके वित्तीय लक्ष्यों और जोखिम उठाने की क्षमता पर निर्भर करता है।\n"
                    "मैं आपको सामान्य जानकारी दे सकता हूँ, लेकिन व्यक्तिगत सलाह के लिए वित्तीय सलाहकार से संपर्क करें\n"
                    "अधिक विशिष्ट जानकारी के लिए कृपया फंड का नाम बताएं (जैसे SBI Bluechip, HDFC Mid-Cap, Parag Parikh Flexi Cap).", None)

    return ("Maaf kijiye, mujhe iski jankari nahi hai. Mai sirf verified mutual funds aur calculations me madad kar sakta hoon.", None)

# --- Flask Routes ---
@app.route('/')
def index():
    return render_template('index.html')

@app.route('/chat_text', methods=['POST'])
def chat_text():
    user_message = request.json.get('message', '')
    response_text, plot_fig = generate_hindi_response(user_message)

    plot_image_base64 = None
    if plot_fig:
        buf = io.BytesIO()
        plot_fig.savefig(buf, format='png', bbox_inches='tight')
        buf.seek(0)
        plot_image_base64 = base64.b64encode(buf.getvalue()).decode('utf-8')

    # Convert bot_response to speech
    tts = gTTS(text=str(response_text), lang='hi') # Added str() cast as a safeguard
    audio_buf = io.BytesIO()
    tts.write_to_fp(audio_buf)
    audio_buf.seek(0)
    audio_base64 = base64.b64encode(audio_buf.getvalue()).decode('utf-8')

    return jsonify({"response_text": response_text, "plot_image_base64": plot_image_base64, "audio_base64": audio_base64})

@app.route('/chat_audio', methods=['POST'])
def chat_audio():
    base64_audio = request.json.get('audio', '')
    text_query = ""
    response_text = ""
    plot_image_base64 = None
    audio_base64_tts = None

    try:
        # The JS MediaRecorder outputs webm, pydub can handle this directly from bytes
        audio_data = base64.b64decode(base64_audio.split(',')[1])
        audio_segment = AudioSegment.from_file(io.BytesIO(audio_data), format='webm')

        # Export to a standard PCM WAV which speech_recognition can reliably handle
        wav_buffer = io.BytesIO()
        audio_segment.export(wav_buffer, format='wav', parameters=['-acodec', 'pcm_s16le', '-ar', '16000'])
        wav_buffer.seek(0)

        recognizer = sr.Recognizer()
        with sr.AudioFile(wav_buffer) as source:
            audio = recognizer.record(source)

        text_query = recognizer.recognize_google(audio, language='hi-IN')
        print(f"\n📢 User (Speech): {text_query}")

        response_text, plot_fig = generate_hindi_response(text_query)

        if plot_fig:
            buf = io.BytesIO()
            plot_fig.savefig(buf, format='png', bbox_inches='tight')
            buf.seek(0)
            plot_image_base64 = base64.b64encode(buf.getvalue()).decode('utf-8')

        tts = gTTS(text=str(response_text), lang='hi') # Added str() cast as a safeguard
        audio_buf = io.BytesIO()
        tts.write_to_fp(audio_buf)
        audio_buf.seek(0)
        audio_base64_tts = base64.b64encode(audio_buf.getvalue()).decode('utf-8')

    except sr.UnknownValueError:
        response_text = "क्षमा करें, मैं आपकी आवाज़ नहीं समझ पाया। कृपया फिर से प्रयास करें।"
    except sr.RequestError as e:
        response_text = f"आवाज़ सेवा से कोई त्रुटि हुई; {e}"
    except Exception as e:
        response_text = f"एक आंतरिक त्रुटि हुई: {e}"
        print(f"Error processing audio: {e}")

    return jsonify({"response_text": response_text, "plot_image_base64": plot_image_base64, "audio_base64": audio_base64_tts})


# --- Manual ngrok setup and Flask execution in thread ---
try:
    # Kill any existing ngrok tunnels to free up resources and prevent conflicts
    ngrok.kill()

    # Set NGROK_AUTH_TOKEN from Colab secrets for stable tunnels
    from google.colab import userdata
    # Check if the secret exists and is not empty
    ngrok_token = userdata.get('NGROK_AUTH_TOKEN')
    if ngrok_token:
        ngrok.set_auth_token(ngrok_token)
    else:
        raise Exception("NGROK_AUTH_TOKEN not found in Colab secrets or is empty. Please add it.")

    public_url = ngrok.connect(5000).public_url
    print(f" * ngrok tunnel available at: {public_url}")
    print(f" * You can access your chatbot at this URL. Remember to adjust your frontend JS to call this backend!")

except Exception as e:
    print(f"Error starting ngrok tunnel: {e}")
    print("Please ensure you have set up your NGROK_AUTH_TOKEN in Colab secrets as instructed above.")


# Run Flask app in a separate thread so Colab doesn't block
def run_flask():
    app.run(port=5000, debug=False, use_reloader=False)

# Check if the Flask thread is already running to prevent "Address already in use"
# This is a basic check and might not be sufficient for all scenarios of repeated execution
# For a more robust solution, especially in Colab, restarting the runtime is often recommended
# or implementing more sophisticated server shutdown logic.
if 'flask_thread' not in globals() or not flask_thread.is_alive():
    flask_thread = threading.Thread(target=run_flask)
    flask_thread.daemon = True # Allow the main program to exit even if the thread is still running
    flask_thread.start()
    print("Flask app started in a separate thread.")
else:
    print("Flask app is already running.")
```
